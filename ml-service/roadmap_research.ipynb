# AI Learning Roadmap Generator - Research & Prompt Engineering

This notebook documents the research and testing phase for the **LearnMate AI++ Roadmap Generator**. 
The goal is to use Gemini API to generate structured learning paths for students.

## 1. Goal Structure
The roadmap should be divided into **Phases** (e.g., Week 1-2, Week 3-4), where each phase contains:
- **Title**: High-level area (e.g., "Foundations of Python")
- **Sub-topics**: Specific skills to learn.
- **Estimated Time**: Duration of the phase.
- **Resources**: Recommended search queries or types of content.

## 2. Prompt Engineering
The prompt needs to ensure JSON output for easy integration with the MERN stack.

```python
topic = "Full Stack Web Development"

prompt = f'''
Generate a professional, detailed learning roadmap for "{topic}".
Divide the roadmap into 4 distinct phases.

Return ONLY a JSON array in this format:
[
  {{
    "phase": "Phase 1: Title",
    "duration": "2 Weeks",
    "topics": ["Topic A", "Topic B", "Topic C"],
    "description": "Brief overview of what will be learned here.",
    "resources": ["Search Query 1", "Resource Type 2"]
  }}
]
No extra text.
'''
print("Prompt Ready for Integration.")
```
